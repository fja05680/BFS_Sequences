```python
# use future imports for python 3.x forward compatibility
from __future__ import print_function
from __future__ import division
from __future__ import absolute_import

# other imports
import random
import math
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib notebook

# project imports
import rand
import halton
import bfs
```

```python
# integration function
def integrate(fn, s, N=100000, m=40, seq=rand.rand_seq):
    ''' return a list of the results of m simulations, i.e.
        m averages of the function fn from 0 to 1
    '''

    theta = []

    for i in range(m):
        nums = seq(s=s, N=N)
        print(len(nums))
        # remove tuple wrapper if necessary
        theta_N = 0
        print('.', end='')
        for j in range(N):
            theta_N += fn(nums[j])
        theta.append(theta_N / N)

    return theta
```

Example:
Pick two points at random within a unit cube. On average, what is the distance between them?  
https://medium.freecodecamp.org/solve-the-unsolvable-with-monte-carlo-methods-294de03c80cd  

```python
fn_cd  = lambda x :  math.sqrt((x[0]-x[1])**2 + (x[2]-x[3])**2 + (x[4]-x[5])**2)
```

```python
#fn = fn_e
#fn = fn_pi
#fn = fn_sq
#fn = fn_sn
fn = fn_cd
#fn = fn_15d
N = 10000
s = 6
m = 40
seq = rand.rand_seq
seq = halton.halton_seq
seq = bfs.bfs_seq
theta = integrate(fn=fn, s=s, N=N, m=m, seq=seq)
```

```python
# print the first 5 estimates of theata
print(theta[:5])
```

```python
# adjust the values for terms left out if necessary
if   fn == fn_e:   theta = [x + 1 for x in theta]
elif fn == fn_pi:  theta = [x * 4 for x in theta]
```

```python
# print the mean and std
theta = numpy.array(theta)
mean = theta.mean()
std = theta.std()
print('mean = {}'.format(mean))
print('standard deviation = {:.2e}'.format(std))
```

```python
# calculate error
if   fn == fn_e:   exact = math.e
elif fn == fn_pi:  exact = math.pi
elif fn == fn_sq:  exact = 7/(6 * math.sqrt(2)) + 1/8 * math.log(3 + 2*math.sqrt(2))
elif fn == fn_sn:  exact = 1/32 * (40 - 8*math.sin(2) - 8*math.sin(4) + math.sin(8))
elif fn == fn_cd:  exact = 0.6617
elif fn == fn_1d:  exact = 5/3
elif fn == fn_2d:  exact = 7/3
elif fn == fn_3d:  exact = 9/3
elif fn == fn_4d:  exact = 11/3
elif fn == fn_5d:  exact = 13/3
elif fn == fn_6d:  exact = 15/3
elif fn == fn_7d:  exact = 17/3
elif fn == fn_8d:  exact = 19/3
elif fn == fn_9d:  exact = 21/3
elif fn == fn_10d: exact = 23/3
elif fn == fn_11d: exact = 25/3
elif fn == fn_12d: exact = 27/3
elif fn == fn_13d: exact = 29/3
elif fn == fn_14d: exact = 31/3
elif fn == fn_15d: exact = 33/3
else:              exact = None

print('exact = {}'.format(exact))
error = abs(mean - exact) / exact * 100
print('error = {:.2e}%'.format(error))
```

```python
# calculate the cummulative mean
cum_sum_theta = numpy.cumsum(theta)
cum_mean_theta = cum_sum_theta / numpy.arange(1, len(cum_sum_theta) +1)
```

```python
# plot the cummulative mean
plt.rcParams['figure.figsize'] = [7, 5]
plt.plot(cum_mean_theta)
plt.ylabel('theta')
plt.show()
```
